```python
# Corr√©lation entre Note Globale et Note du Meilleur √âpisode
corr = df['Note_Globale'].corr(df['Note_Meilleur_Ep'])

print(f"üîó Corr√©lation Note Globale vs Note Meilleur Ep : {corr:.2f}")

# Interpr√©tation
if corr > 0.5:
    print("\nüìà Corr√©lation POSITIVE forte !")
    print("   ‚Üí Un bon √©pisode = souvent un bon anim√©")
elif corr > 0:
    print("\nüìà Corr√©lation POSITIVE faible")
else:
    print("\nüìâ Corr√©lation N√âGATIVE")
```

```python
# Corr√©lation Note Globale vs Note du Pire √âpisode
corr_pire = df['Note_Globale'].corr(df['Note_Pire_Ep'])

print(f"üîó Corr√©lation Note Globale vs Note Pire Ep : {corr_pire:.2f}")

# Comparaison
print(f"\nüìä R√©capitulatif :")
print(f"   ‚Ä¢ Note Meilleur Ep : forte corr√©lation positive")
print(f"   ‚Ä¢ Note Pire Ep    : corr√©lation plus faible")
print(f"\nüí° Un anim√© peut avoir un mauvais √©pisode et rester bien not√© !")
```

```python
# On calcule l'√©cart entre le plafond (Top) et le plancher (Flop)
df['Ecart_Qualite'] = df['Note_Meilleur_Ep'] - df['Note_Pire_Ep']
```

```python
# --- √âTAPE H3 : CONSTRUCTION DU SCORE (Solution) ---

# On applique la p√©nalit√© de r√©gularit√©
# Formule : Note Globale - (10% de l'√©cart)
df['Score_Editorial'] = df['Note_Globale'] - (df['Ecart_Qualite'] * 0.1)

# VERIFICATION VISUELLE
# On regarde les anim√©s qui ont le plus gros √©cart pour voir si leur note a baiss√©
colonnes = ['Anime', 'Note_Globale', 'Ecart_Qualite', 'Score_Editorial']
print("üëÄ Top 5 des anim√©s les plus p√©nalis√©s (Irr√©guliers) :")
display(df.sort_values('Ecart_Qualite', ascending=False)[colonnes].head())
```

```python
# --- √âTAPE H4 : SEGMENTATION & D√âCISION (Le Final) ---

# On cr√©e la fonction qui donne les m√©dailles
def segmenter_anime(row):
    # La P√âPITE : Super Note (>8.5) ET Tr√®s Stable (Ecart < 2)
    if row['Score_Editorial'] >= 8.5 and row['Ecart_Qualite'] < 2:
        return "üíé P√âPITE (A acheter)"
    
    # La VALEUR S√õRE : Bonne note (>7.5)
    elif row['Score_Editorial'] >= 7.5:
        return "‚úÖ Valeur S√ªre"
    
    # LE RISQUE : Trop instable (Ecart > 4)
    elif row['Ecart_Qualite'] > 4:
        return "‚ö†Ô∏è Risqu√© (Irr√©gulier)"
    
    # LE RESTE
    else:
        return "‚ùå Fond de catalogue"

# On applique la fonction √† tout le monde
df['Decision_Finale'] = df.apply(segmenter_anime, axis=1)

# --- EXPORT DU FICHIER ---
# On trie par ton Score pour avoir les meilleurs en haut
df_final = df.sort_values('Score_Editorial', ascending=False)
colonnes_utiles = ['Anime', 'Note_Globale', 'Ecart_Qualite', 'Score_Editorial', 'Decision_Finale']

print("üéâ PROJET TERMIN√â ! Voici tes champions :")
display(df_final[colonnes_utiles].head(10))

# Sauvegarde pour l'envoi au prof
df_final.to_csv('resultat_final_projet.csv', index=False)
print("‚úÖ Fichier 'resultat_final_projet.csv' pr√™t √† √™tre envoy√© !")
```
